# Employee Attrition Prediction - Exploratory Data Analysis (EDA)

This notebook performs step-by-step exploratory data analysis (EDA) on the IBM HR Analytics Attrition Dataset.

```python
# Step 1: Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
%matplotlib inline
```

## Step 2: Load Dataset

```python
# Read the dataset
df = pd.read_csv("../data/WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head()
```

## Step 3: Dataset Overview

```python
# Shape and column names
print("Rows, Columns:", df.shape)
df.columns
```

```python
# Basic info and datatypes
df.info()
```

## Step 4: Check for Missing Values and Duplicates

```python
# Missing values
df.isnull().sum().sort_values(ascending=False)
```

```python
# Check duplicates
df.duplicated().sum()
```

## Step 5: Target Variable Distribution

```python
sns.countplot(x='Attrition', hue='Attrition', data=df, palette='Set2', legend=False)
plt.title("Distribution of Employee Attrition")
plt.show()
```

## Step 6: Numerical Feature Analysis

```python
# Summary statistics
df.describe()
```

```python
# Copy to avoid mutating the original
df_corr = df.copy()

# Auto-detect and map all Yes/No columns to 1/0 (including Attrition)
yes_no_cols = [
    c for c in df_corr.select_dtypes(include='object').columns
    if df_corr[c].dropna().isin(['Yes', 'No']).all()
]
for c in yes_no_cols:
    df_corr[c] = df_corr[c].map({'Yes': 1, 'No': 0})

# Compute correlation on numeric columns (now includes Attrition)
num_cols = df_corr.select_dtypes(include=['number']).columns
corr = df_corr[num_cols].corr()

plt.figure(figsize=(14, 10))
sns.heatmap(corr, annot=False, cmap='coolwarm')
plt.title("Correlation Heatmap (with Yes/No mapped to 1/0)")
plt.show()
```

### (Optional) Focus on features most related to Attrition

#### This makes the plot easier to read:

```python
# Ensure 'Attrition' exists after mapping
if 'Attrition' in corr.columns:
    top = corr['Attrition'].abs().sort_values(ascending=False).head(15).index
    plt.figure(figsize=(10, 8))
    sns.heatmap(df_corr[top].corr(), annot=True, fmt=".2f", cmap='coolwarm')
    plt.title("Top Correlations with Attrition")

    # Save the plot into visuals folder
    plt.savefig("../visuals/correlation_heatmap.png", dpi=300, bbox_inches="tight")

    plt.show()
else:
    print("'Attrition' not found in correlation columns. Check mapping step.")
```

```python
# Correlation Heatmap
plt.figure(figsize=(14,10))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

## Step 7: Categorical Features vs Attrition

```python
# Attrition by Gender
sns.countplot(x='Gender', hue='Attrition', data=df)
plt.title("Attrition by Gender")
plt.show()
```

```python
# Attrition by Department
plt.figure(figsize=(8,5))
sns.countplot(x='Department', hue='Attrition', data=df)
plt.title("Attrition by Department")
plt.show()
```
